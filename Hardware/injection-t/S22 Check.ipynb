Check Reflection Phase for Lowpass and Highpass filters to determine suitability for use in diplexer. We're interested in S22

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
#LFCN-490D+
lfcn_490d = np.loadtxt("LFCN-490D+_S2P/LFCN-490D+_UNIT 1.S2P",
                      skiprows = 13)

fig1 = plt.figure(figsize = (9,5))

plt.plot(lfcn_490d[:2000,0], lfcn_490d[:2000,5])
plt.plot(lfcn_490d[:2000,0], lfcn_490d[:2000,8])
plt.ylabel("SXX (dB)")
plt.xlabel("Frequency (Hz)")
plt.show()
```

```python
#HFCN-440+
hfcn_440 = np.loadtxt("HFCN-440+_S2P/HFCN-440+___Plus25degC.S2P",
                      skiprows = 9)

fig2 = plt.figure(figsize = (9,5))

plt.plot(hfcn_440[:,0], hfcn_440[:,7])
plt.plot(hfcn_440[:,0], hfcn_440[:,8])
plt.ylabel("SXX (dB)")
plt.xlabel("Frequency (Hz)")
plt.show()
```
